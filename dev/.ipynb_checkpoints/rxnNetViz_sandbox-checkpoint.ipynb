### Import dependencies

```julia
include("../src/AutoChem.jl")
include("../src/config.jl")
include("../src/parse_mcm.jl")

include("../src/parse_mcm.jl")
```

```julia
# using AutoChem
test_file = "../data/test.kpp"
isfile(test_file)

out_file = "../data/test.jl"
isfile(out_file)
```

### 1. Read in mechanism

```julia
test_file
rxns, species_list = extract_mechanism(test_file)

for rxn ∈ rxns
    println("Rxn: $(rxn[1]),  Rate: $(rxn[6])")
end
```

### 2. Generate ReactionNetwork file

```julia
mechanism_to_catalyst(rxns, species_list, out_file, "test_rn")
```

### 3. Load in reaction network

```julia
include(out_file)

@variables t, A(t), B(t), C(t)
test_rn
rn_test = @reaction_network begin
    1.0, A + B → C
end

test_rn == rn_test

display(test_rn)
display(rn_test)

isequal(test_rn, rn_test)

rx_a = reactions(test_rn)
a_rx = reactions(rn_test)

rx_a == a_rx
```

### 4. Visualize

```julia
g = Graph(test_rn)
display(g)
```
